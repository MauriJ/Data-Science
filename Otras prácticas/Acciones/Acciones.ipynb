```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
names = ['D', 'DUK', 'EXC', 'NEE', 'SO']

list_df = []

for i in names:
    df = pd.read_csv(f'{i}.csv')
    df['Symbol'] = i # Agregar nueva columna con el simbolo 
    list_df.append(df)
    
df_total = pd.concat(list_df, axis=0)
```

```python
df_total.head()
```

```python
df_total.hist(bins=30, figsize=(10, 6))
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data = df_total, x = 'Symbol', y = 'Open')
plt.show()
```

```python
pd.plotting.scatter_matrix(df_total, figsize=(10,10), marker='.', hist_kwds={'bins': 20}, s=20)
plt.show()
```

```python
df_total.corr().round(3)
```

```python
sns.heatmap(df_total.corr().round(2), vmin=-1, vmax=1, cbar = True, annot=True)
plt.show()
```
