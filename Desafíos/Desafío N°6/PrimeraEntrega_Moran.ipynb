### - Definición de objetivo: 

El objetivo es crear un sistema de clasificación de canciones en función de diversos parámetros, tales como género, energía que transmite, la popularidad, la acústica, etcétera. ### - Contexto comercial:

El servicio Spotify es un servicio por suscripción, es decir, los clientes pagan mensualmente la suscrpición que le da acceso al servicio, gozando beneficios tales como acceso a las canciones, podcasts, reproducción offline, etcétera. ### Adquisición de datos

Los datos fueron obtenidos a partir de un archivo .csv obbtenido de la plataforma Kaggle. Este dataset fue construido a partir de datos extraídos de Spotify Web API. El dataset está compuesto por una muestra de 114.000 canciones, la cual cada una de ellas es descripta por 21 variables. Entre dichas variables se encuentra el nombre de la canción, su intérprete, álbum, duración y una serie de parámetros numéricos que dan información sobre el volumen, la energía que transmite, el carácter instrumental, la presencia de palabra hablada o cantanda, entre otras. A continuación se describen cada variable y su significado.

- Unnamed: 0: Número de instancia.
- track_id: Número identificatorio de la canción.
- artists: Nombre de los artistas.
- album_name: Nombre del album al cual la canción pertenece.
- track_name: Nombre de la canción.
- popularity: Popularidad de la canción, toma valores entre 0 y 100, siendo 100 el más popular. 
- duration_ms: Duración de la canción en milisegundos.
- explicit: Informa si la canción posee letras explícitas (True) o no (False).
- danceability: Describe cuan bailable es una canción, basado en una combinación de elementos musicales tales como tempo, ritmo, etcétera. Valor 0.0 indica poco bailable y 1.0 indica muy bailable.
- energy: Energy es una medida entre 0.0 y 1.0 y representa una medida de la intensidad y la actividad. Típicamente canciones energéticas se sienten rápidas, de volumen alto y ruidosas.
- key: La tonalidad de la pista. Los números enteros se asignan a tonos utilizando la notación estándar de clase de tono. Por ejemplo, 0 = C, 1 = C♯/D♭, 2 = D, y así sucesivamente. Si no se detecta ninguna clave, el valor es -1.
- loudness: Volumen promedio de la canción en decibelios (dB).
- mode: El modo indica la modalidad (mayor o menor) de una pista, el tipo de escala del que se deriva su contenido melódico. Mayor se representa con 1 y menor con 0.
- speechiness: Speechiness detecta la presencia de palabra hablada. Mientras más palabra hablada posea (talk show, audiobook, poesía), más cerca de 1 estará el valor. Valores por debajo de 0.33 representan música y otros temas con voz cantada.
- acousticness: Medida desde 0.0 a 1.0 sobre si la canción es acústica, indicando 1.0 que la canción es acústica.
- instrumentalness: Predice si la canción no contiene voces. Mientras más cerca esté de 1.0, mayor es la posibilidad que la canción no contenga voces.
- liveness: Detecta la presencia de audiencia en las canciones. Altos valores indican alta posibilidad de que el tema se interpretó en vivo.
- valence: Medida desde 0.0 a 1.0 describiendo la positividad musical que la canción porta. Canciones con alta valencia suenan más positivas (felices, animadas, eufóricas), mientras la baja valencia indica temas más negativos (e.g. tristeza, depresivo, ira).
- tempo: El tempo promedio de una canción en pulsaciones (beats) por minuto (BPM). En terminología musical, el tempo es la velocidad o el ritmo de una pieza determinada y se deriva directamente de la duración media del compás.
- time_signature: Compás estimado, es una convención de notación que especifica cuántos tiempos hay en cada compás. El compás oscila entre 3 y 7, lo que indica compases de 3/4 a 7/4.
- track_genre: Género al que la canción pertenece.

### Géneros, artistas y canciones más populares

```python
popular = df['popularity'].groupby(df['artist_1']).mean()
popular.sort_values(ascending = False).head(5)
```

```python
df.sort_values(by = ['popularity'], ascending = False).head(10)
```

Se observa que las canciones más populares están dominadas por géneros pop, hip-hop y latino, canciones que altos valores de energy y danceability. Indicando que posiblemente las canciones con mayor predisposición a ser bailables son las más exitosas. 

Es crucial destacar que muchas de las canciones se encuentran repetidas varias veces, encontrando que fueron clasificadas en géneros distintos. Esta repetición debe ser analizada con conciencia puesto que puede afectar negativamente a los algoritmos de aprendizaje automático.

### Datos duplicados

```python
df.duplicated(subset = ['track_name']).value_counts()
```

El total de canciones repetidas asciende a 40.391, de un total de 114.000, lo cual es el 35% del dataset, un número francamente desproporcionado. Muy probablemente estas canciones repetidas deban ser eliminadas antes de entrenar los algoritmos de aprendizaje automático.

### Duración promedio de las canciones

```python
sns.histplot(data = df, x = 'duration_ms', color = '#1DB954')
plt.xlim(0, 1.25e6)
plt.xlabel('Duración (en ms)', fontsize = 14) 
plt.ylabel('Conteo', fontsize = 14) 
plt.title('Distribución de la duración de las canciones', fontsize = 16)
plt.show()
```

El centro de la distribución (moda) se encuentra en torno a los 0.2E6 milisegundos, un poco más de cuatro minutos, sin embargo hay canciones que superan holgadamente ese valor. 

### Letras explícitas

```python
plt.figure(figsize = (4,4))
_, _, autotexts = plt.pie(df.groupby('explicit').size(), labels = ('No contiene', 'Contiene'), explode = (0.1, 0), startangle = 45, autopct = '%1.1f%%', colors = ['#1DB954', '#191414'], wedgeprops = {"edgecolor" : "black", 'linewidth': 1})
for ins in autotexts:
    ins.set_color('white')

plt.title('Letras explícitas', fontsize = 16)
plt.show()
```

Se encuentra que el 8.6 % de las canciones poseen letras explíctas. Esta es otra variable a considerar, filtrar las canciones no explícitas de las explícitas es de utilidad para padres que buscan no exponer a sus hijos de corta edad a letras no aptas para menores.

### Analisis de las variables entre géneros musicales

```python
df_seleccion = df[df['track_genre'].isin(['classical', 'opera', 'rock', 'hard-rock', 'pop', 'k-pop', 'dance', 'disco'])]

columns = df_seleccion.columns[4:-3] # Selecciono los parámetros que considero más relevantes

l = [i for i in columns if i not in ['explicit', 'mode']]

plt.figure(figsize = (14, 18))
      
for x, column in zip(range(12), df_seleccion[l]):
    
    plt.subplot(int(len(l)/2), 2, x + 1)
    sns.boxplot(data = df_seleccion, y = column, x = df_seleccion['track_genre'], color = '#1DB954', showfliers = False) 
    # Filtro los outliers porque distorsionan los gráficos, dificultando apreciar las diferencias
    
    plt.xticks(rotation = 25)
    plt.xlabel("")
    plt.ylabel(l[x].capitalize(), fontsize = 14)
    
plt.tight_layout()
```

En estos gráficos de cajas se puede observar que los distintos géneros presentan diferencias sustanciales entre las variables, como se mencionó en la entrega anterior esto es el primer argumento para hacer emplear un algoritmo de clasificación para armar el sistema de recomendación de canciones. Por otro lado, si comparamos entre los pares "classical" - "opera", "dance" - "disco", "rock" - "hard-rock" y "pop" - "k-pop", notamos que presentan cierta similaritud por lo que factible que un cliente que escuche ópera posiblemente le guste escuchar música clásica. Esto permite que el sistema de recomendación tenga cierto márgen para ofrecer géneros nuevos pero similares a los que la audiencia ya escuche. 

### Analisis de distribución de datos y de correlaciones entre pares de variables

```python
df.iloc[:, 1:].hist(bins = 30, figsize = (15, 10), color = '#1DB954')
plt.tight_layout()
plt.show()
```

Los histogramas de las variables numéricas del dataset muestran un gran abanico de distribuciones, es por esto que deberá tenerse especial cuidado a la hora de plantear la estandarización de estos datos. Variables como "tempo" y "danceablility" ya tienen un comportamiento similar al normal o distribuciones similares a log-normal ("loudness", "liveness") y otras variables presentan valores discretos ("mode", "time_signature"). Existen numerosos algoritmos de estandarización y habrá que estudiarlos y evaluarlos antes de proceder.

```python
df1 = df.drop(['Unnamed: 0', 'key', 'explicit', 'mode', 'time_signature'], axis = 1) 

pd.plotting.scatter_matrix(df1.iloc[::100,:], color = '#1DB954', figsize=(14,14), marker='.', hist_kwds={'color':['#1DB954'], 'bins': 20}, s=20, alpha =.8)
plt.tight_layout()
plt.show()
```

El gráfico anterior muestra gráficos de puntos entre todos los pares de variables numéricas presentes en el dataset. Este gráfico es voluminoso y complicado de extraer conclusiones, por lo que se repite este gráfico pero con un número de variables más reducido, se seleccionaron aquellas que presentan cierta correlación.

```python
df2 = df[['danceability', 'energy', 'loudness', 'acousticness', 'instrumentalness', 'valence']]

pd.plotting.scatter_matrix(df2.iloc[::100,:], color = '#1DB954', figsize=(10,10), marker='.', hist_kwds={'color':['#1DB954'], 'bins': 20}, s=20, alpha =.8)
plt.tight_layout()
plt.show()
```

Este gráfico muestra resumidamente pares de variables que presentan correlaciones de distinto grado, algunas presentan fuerte correlación con lo que se podría pensar en disminuir el número de variables a la hora de entrenar el algoritmo de clasificación. Sin embargo, en otros casos la correlación es débil, indicando cierta vinculación entre ellas pero que no ameritaría la redución de una variable.
