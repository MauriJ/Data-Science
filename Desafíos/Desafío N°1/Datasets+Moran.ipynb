## Database #1: Spotify tracks dataset

#### Fuente: https://www.kaggle.com/datasets/maharshipandya/-spotify-tracks-dataset

```python
df_s = pd.read_csv('Spotify_Track_Dataset.csv')
```

```python
df_s.head()
```

```python
df_s.info()
```

#### El archivo está compuesto por 114.000 canciones incluidas en el servicio Spotify, la información de cada una de esas canciones se encuentra distribuida en 21 columnas. Estas detallan el nombre de las canciones, los interpretes, el album del que forma parte, su duración, etcétera. El archivo también incluye columnas tales como "danceability", "energy", "speechiness", "acousticness", entre otros, que dotan de valores numéricos (generalmente entre 0 y 1) a dichos parámetros categóricos y subjetivos (más información en el link). Estas variables son útiles para caracterizar a grandes rasgos el enorme número de canciones disponibles en Spotify, con lo cual podríamos desarrollar un sistema de agrupamiento de canciones a partir de sus características en común. Esto permitiría desarrollar un sistema de calificación y recomendación de canciones a los usuarios en función de sus gustos.

```python
round(df_s.describe(), 2)
```

```python
df_s.isna().sum()
```

#### En la celda inmediatamente anterior se verificó el número de entradas vacías o NaN en cada columna, podemos observar que sólo tres columnas poseen sólo un valor de NaNs. Esto es sumamente positivo, indicando que muy probablemente el dataset se encuentre completo.  

## Database #2: Air traffic passenger statistics

#### Fuente: https://www.kaggle.com/datasets/thedevastator/airlines-traffic-passenger-statistics

```python
df_at = pd.read_csv('Air_Traffic_Passenger_Statistics.csv')
```

```python
df_at.head()
```

```python
df_at.info()
```

#### Este dataset contiene información sobre el tráfico aéreo de transporte de pasajeros del 2017. Este brinda información sobre los distintos vuelos (15007 entradas), tales como nombre de la aerolínea, fecha del vuelo, número de pasajeros, origen y destino del vuelo, precio del vuelo, entre otras columnas (17 columnas en total). Estos datos podrían ser usados para estudiar el patrón de vuelos realizados en ese año ser usados para predecir los vuelos en años futuros. También se podría analizar el efecto de variables tales como el momento del año, el precio del vuelo, entre otras variables.

```python
round(df_at.describe(), 2)
```

```python
df_at.isna().sum()
```

#### Se verifica la presencia de valores NaN en el dataset, encontrándolos sólo en las columnas Airline IATA Code. Estas podrían ser posiblemente subsanadas con facilidad, puesto que los código IATA son claves constituidas por dos caracteres alfanuméricos, con las cuales se identifican las aerolíneas (https://en.wikipedia.org/wiki/Airline_codes).
